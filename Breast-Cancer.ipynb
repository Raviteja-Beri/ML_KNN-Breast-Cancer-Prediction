```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop(['Class'], axis=1)

y = data['Class']
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize kNN with 7 neighbors
knn_7 = KNeighborsClassifier(n_neighbors=7)

# Train model on scaled data
knn_7.fit(X_train_scaled, y_train)
```

```python
y_pred = knn_7.predict(X_test)

y_pred
```

```python
knn_7.predict_proba(X_test)[:,0]
```

```python
knn_7.predict_proba(X_test)[:,1]
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# Generate predictions for ROC curve
y_pred_1 = knn_7.predict(X_test_scaled)
```

```python
from sklearn.metrics import roc_curve
```

```python
y_pred_proba = knn_7.predict_proba(X_test_scaled)[:, 1]
```

```python
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label=4)

plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.rcParams['font.size'] = 12
plt.title('ROC curve for Breast Cancer kNN classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred_proba)
print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

```python
from sklearn.model_selection import cross_val_score

# Cross-validate ROC AUC using probability scores
Cross_validated_ROC_AUC = cross_val_score(knn_7, X_train_scaled, y_train, cv=5, scoring='roc_auc').mean()
print('Cross validated ROC AUC : {:.4f}'.format(Cross_validated_ROC_AUC))
```

```python
scores = cross_val_score(knn_7, X_train_scaled, y_train, cv=10, scoring='accuracy')
print('Cross-validation scores:{}'.format(scores))
```

```python
print('Average cross-validation score: {:.4f}'.format(scores.mean()))
```

```python
# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
```

```python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

```python
true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```
